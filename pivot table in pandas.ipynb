```python
import pandas as pd
df = pd.DataFrame({"Student_name": ["Kyle", "Max", "Kyle", "Max", "Chris",
                         "Tom", "Diana", "Max", "Kyle"],
                   "Subject":["Math", "Physics", "English", "English", "Math", "Math","Math", "Physics", "Physics"],
                    "Score":[80, 90, 67, 56, 66, 88, 93,92, 54]})
```

```python
df
```
